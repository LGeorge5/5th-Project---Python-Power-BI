```python
def Get_Income_Cat(income):
    if income <= 3847.5:
        return 'Low income'
    elif income <= 12000.0:
        return 'Medium income'
    elif income <= 60000.0:
        return 'High income'
    else:
        return 'Great income'
print(Get_Income_Cat(79999))
```

```python
print(Get_Income_Cat(11999))
```

```python
df['IncomeCat']=df['income'].apply(Get_Income_Cat)
df.head(3)
```

```python
df['countries_shipped_to'].describe()
```

```python
def Get_CoShip_Cat(coship):
    if coship <= 32:
        return '32 countries or lower'
    elif coship <= 40:
        return 'Between 33 and 40 countries'
    elif coship <= 43:
        return 'Between 41 and 43 countries'
    else:
        return '44 countries or higher'
print(Get_CoShip_Cat(125))
```

```python
print(Get_CoShip_Cat(38))
```

```python
df['coship_cat']=df['countries_shipped_to'].apply(Get_CoShip_Cat)
df.tail(3)
```

```python
def Get_MeRating_Cat(mrate):
    if mrate <= 2:
        return 'Low Rating'
    elif mrate <= 3:
        return 'Medium Rating'
    elif mrate <= 4:
        return 'High Rating'
    else:
        return 'Great Rating'
print(Get_MeRating_Cat(3.5))
```

```python
df['MeRating_Cat']=df['merchant_rating'].apply(Get_MeRating_Cat)
df.tail(3)
```

```python
df['Rating_Cat']=df['rating'].apply(Get_MeRating_Cat)
df.tail(3)
```

```python
df['price_difference']=df['retail_price']-df['price']
df.head(3)
```
